# 02 Exploratory Analysis

This notebook looks at profit, delivery reliability, customer behaviour, and discount patterns.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/processed/food_delivery_cleaned.csv')
```

```python
kpis = {'orders': len(df), 'gross_profit': round(df['gross_profit'].sum(),2), 'avg_order_value': round(df['order_value'].mean(),2), 'repeat_rate_pct': round(df['repeat_customer_flag'].mean()*100,2), 'delay_rate_pct': round(df['delivery_delay_flag'].mean()*100,2)}
kpis
```

```python
df.groupby('city').agg(orders=('order_id','count'), gross_profit=('gross_profit','sum'), avg_margin=('profit_margin_pct','mean'), delay_rate=('delivery_delay_flag','mean'), repeat_rate=('repeat_customer_flag','mean')).sort_values('gross_profit', ascending=False)
```

```python
df.groupby('discount_band').agg(orders=('order_id','count'), avg_discount=('discount_amount','mean'), avg_profit=('gross_profit','mean'), repeat_rate=('repeat_customer_flag','mean'), avg_rating=('rating','mean'))
```

```python
df.groupby(['city','zone']).agg(orders=('order_id','count'), delay_rate=('delivery_delay_flag','mean'), avg_delay=('delay_minutes','mean'), avg_rating=('rating','mean')).query('orders >= 30').sort_values(['delay_rate','avg_rating'], ascending=[False, True]).head(12)
```
