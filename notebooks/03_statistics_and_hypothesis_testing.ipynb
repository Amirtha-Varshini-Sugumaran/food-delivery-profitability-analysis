# 03 Statistics And Hypothesis Testing

The goal is practical evidence, not academic complexity.

```python
import pandas as pd
from scipy import stats

df = pd.read_csv('../data/processed/food_delivery_cleaned.csv')
```

## Are delayed orders associated with lower ratings?

```python
delayed = df.loc[df['delivery_delay_flag'] == 1, 'rating']
on_time = df.loc[df['delivery_delay_flag'] == 0, 'rating']
test = stats.ttest_ind(delayed, on_time, equal_var=False)
{'avg_rating_delayed': round(delayed.mean(),2), 'avg_rating_on_time': round(on_time.mean(),2), 'difference': round(on_time.mean()-delayed.mean(),2), 'p_value': round(test.pvalue,6)}
```

## Do discounts and repeat behaviour show an association?

```python
contingency = pd.crosstab(df['discount_band'], df['repeat_customer_flag'])
chi2, p, dof, expected = stats.chi2_contingency(contingency)
repeat_by_discount = df.groupby('discount_band').agg(repeat_rate=('repeat_customer_flag','mean'), avg_profit=('gross_profit','mean'), avg_discount=('discount_amount','mean'))
round(p, 6), repeat_by_discount
```

## Is profitability different by tenure group?

```python
groups = [g['gross_profit'].values for _, g in df.groupby('customer_tenure_group')]
f_stat, p_value = stats.f_oneway(*groups)
tenure_profit = df.groupby('customer_tenure_group')['gross_profit'].agg(['count','mean','median'])
round(p_value, 6), tenure_profit
```
